## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
df["subtotal"] = df["qty"] * df["unit_price"]
df.info()
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
df["total_weight"] = df["qty"] * df["unit_weight"]
def calc_shipping_price(x):
    if x >50:
        return x*7
    else:
        return x*10
    
df["shipping_price"] = df["total_weight"].apply(calc_shipping_price)
df[["unit_weight","qty","total_weight","shipping_price"]].head()
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
df["tax_rate"] = 0.0925
df["total_tax"] = ((df["subtotal"] + df["shipping_price"]) * df["tax_rate"]).round(2)
df["total_price"] = df["subtotal"] + df["shipping_price"] + df["total_tax"]
df[["subtotal","shipping_price","tax_rate","total_tax","total_price"]].head()
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).
df["total_cost"] = (df["qty"] + df["unit_cost"]) + df["shipping_price"]
df[["qty","unit_cost","shipping_price","total_cost"]].head()
```

```python
# Create a column for the profit of each line using line cost and line price
""" Profits would also deduct the collected sales taxes.
"""

df["profit"] = df["total_price"] - (df["total_cost"] + df["total_tax"])
df["profit"].head()
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.
""" This is the top five clients by the number of times that client appears in 
the data frame, NOT the top clients by quantity.
"""

for i in topfiveclients:
    spend = df.loc[df["client_id"] == i]["total_price"].sum().round(2)
    print(f"Client ID {i} spent ${spend:,.2f}")


# topcategory = list(df["category"].value_counts().head(1).index)
# df.loc[df["category"]==topcategory[0]]["subcategory"].value_counts().head(1)
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit.
""" Taxes are not considered revenues, nor are taxes included in profits.
""" 

df["revenue"] = df["total_price"] - df["total_tax"]

new_df_columns = ["qty","shipping_price","revenue","profit"]
new_client_dict = {"client_id": topfiveclients}

for e in new_df_columns:
    new_client_dict[e] = [df.loc[df["client_id"] == i][e].sum().round() for i in topfiveclients]

new_df = pd.DataFrame(new_client_dict)
new_df
```

```python
# Format the data and rename the columns to names suitable for presentation.

# Define the money columns.
money_columns = ["shipping_price","revenue","profit"] 


# Define a function that converts a dollar amount to millions.
def currency_format_millions(m):
    return m/1000000

# Apply the currency_format_millions function to only the money columns. 
new_df[money_columns] = new_df[money_columns].apply(currency_format_millions).round(2)
new_df

# Rename the columns to reflect the change in the money format.
new_df_clean = new_df.rename(columns={"client_id":"Client ID Number" , "qty":"Quantity Purchased" , "shipping_price":"Total Shipping Price (millions)" , "revenue":"Total Revenue (millions)" , "profit":"Total Profit (millions)"})
new_df_clean
```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.
new_df_sort = new_df_clean.sort_values("Total Profit (millions)",ascending=False)
new_df_sort
```
